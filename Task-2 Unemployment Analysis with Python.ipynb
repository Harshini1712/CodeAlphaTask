```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv(r"C:\Users\User\Downloads\Unemployment in India (1).csv")
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Handle missing values
data.dropna(inplace=True)
data
```

```python
# Summary statistics
print(data.describe())
```

```python
# Convert 'Date' column to datetime
data[' Date'] = pd.to_datetime(data[' Date'].str.strip(), format='%d-%m-%Y')
data[' Date']  
```

```python
# Line plot of unemployment rate over time for a specific region (e.g., Tamil Nadu)
Tamil_Nadu_data = data[data['Region'] == 'Tamil Nadu']

plt.figure(figsize=(12, 6))
sns.lineplot(x=' Date', y=' Estimated Unemployment Rate (%)', data=Tamil_Nadu_data, hue='Area')
plt.title('Unemployment Rate Over Time in Tamil Nadu')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.show()
```

```python
# lineplot of unemployment rate by region
plt.figure(figsize=(12, 8))
sns.lineplot(x='Region', y=' Estimated Unemployment Rate (%)', data=data, hue='Area')
plt.title('Boxplot of Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(data=data, x='Area')
plt.title('Distribution of Area')
plt.show()
```

```python
sns.histplot(data=data, x=' Estimated Unemployment Rate (%)', kde=True, hue= 'Area')
plt.title('Distribution of  Estimated Unemployment Rate (%)')
plt.show()
```

```python
sns.histplot(data=data, x=' Estimated Employed', kde=True, hue = 'Area')
plt.title('Distribution of  Estimated Employed')
plt.show()
```

```python
data[' Date']=pd.to_datetime(data[' Date'],dayfirst=True)
```

```python
data['month_int']=data[' Date'].dt.month
data.head()
```

```python
data.columns= ["Region"," Date"," Frequency",
               " Estimated Unemployment Rate (%)"," Estimated Employed",
               " Estimated Labour Participation Rate (%)","Region.1",
               "longitude","latitude","month_int"]
plt.title("Indian Unemployment")
sns.histplot(x=" Estimated Employed", hue="Region.1", data=data)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
plt.title("Indian Unemployment")
sns.histplot(x=" Estimated Unemployment Rate (%)", hue="Region.1", data=data)
plt.show()
```

```python
after_lockdown = data[(data['month_int'] >= 4) & (data['month_int'] <=6)]
before_lockdown = data[(data['month_int'] >= 1) & (data['month_int'] <=4)]
```

```python
af_lockdown=after_lockdown.groupby('Region')[' Estimated Unemployment Rate (%)'].mean().reset_index()
lockdown= before_lockdown.groupby('Region')[' Estimated Unemployment Rate (%)'].mean().reset_index()
lockdown['unemployment rate before lockdown'] = af_lockdown[' Estimated Unemployment Rate (%)']
lockdown.columns=['state','unemployment rate before lockdown','unemployment rate after lockdown']
lockdown
```
